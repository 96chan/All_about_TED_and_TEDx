```python
import pandas as pd
from pandas import Series, DataFrame
import json

TEDX_JSON_FILE = "../data/final_datamining_ted.json"
ted_video_list = []
with open(TEDX_JSON_FILE, "r") as ted_json_file:
    for line in ted_json_file:
        ted_video_list.append(json.loads(line))
    ted_df = DataFrame(ted_video_list)
ted_df
```

```python
df = pd.DataFrame(ted_df, columns  = ['viewCount'])

filter_values = []

for i in range(1000):
    filter_values.append(i*100)
    
out = pd.cut(df.viewCount, bins = filter_values)
counts = pd.value_counts(out)
#print counts
# counts is a Series
print(counts.sort_index())
```

```python
counts.sort_index()[:30].plot(kind="bar",use_index=True,  figsize=(20,8))
```
