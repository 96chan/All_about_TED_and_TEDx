```python
import pandas as pd
from pandas import Series, DataFrame
import json

TEDX_JSON_FILE = "../data/final_datamining_ted.json"
ted_video_list = []
with open(TEDX_JSON_FILE, "r") as ted_json_file:
    for line in ted_json_file:
        ted_video_list.append(json.loads(line))
    ted_df = DataFrame(ted_video_list)
ted_df
```

```python
df = pd.DataFrame(ted_df, columns  = ['viewCount'])

filter_values = []

for i in range(1000):
    filter_values.append(i*100)
    
out = pd.cut(df.viewCount, bins = filter_values)
counts = pd.value_counts(out)
#print counts
# counts is a Series
tmp_counts = []
for i in range(0, len(counts)):
    index = counts.index[i]
    tmp_counts.append((int(index[1:index.find(',')]), counts[i]))

from operator import itemgetter, attrgetter
tmp_counts = sorted(tmp_counts, key=itemgetter(0))
print tmp_counts
counts_index = list(x[0] for x in tmp_counts)
counts_value = list(x[1] for x in tmp_counts)
counts = Series(counts_value, index=counts_index)
counts[:100].plot(kind="bar",use_index=True,  figsize=(20,8))
#counts.index[0]
#count_df = pd.DataFrame(counts)
#new_count_df = count_df.sort_index()
#new_count_df[0]
```
