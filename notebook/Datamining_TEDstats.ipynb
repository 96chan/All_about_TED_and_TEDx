```python
import pandas as pd
from pandas import Series, DataFrame
import json

TEDX_JSON_FILE = "../data/final_datamining_ted.json"
ted_video_list = []
with open(TEDX_JSON_FILE, "r") as ted_json_file:
    for line in ted_json_file:
        ted_video_list.append(json.loads(line))
    ted_df = DataFrame(ted_video_list)
```

```python
from operator import itemgetter, attrgetter

INTERESTS_VALUE = ['viewCount', 'FB_like_count', 'FB_share_count', 'FB_click_count', 'twitter_count','FB_comment_count', 'commentCount', 'ratingCount', 'duration', 'favoriteCount', 'likeCount']
COUNTS_VALUE = []

for interested in INTERESTS_VALUE:
    df = pd.DataFrame(ted_df, columns  = [interested])
    
    filter_values = []
    SIZE_BUCKET = 100 if interested == 'viewCount' or interested == 'duration' else 10
    
    for i in range(1000):
        filter_values.append(i*SIZE_BUCKET)
        
    out = pd.cut(df[interested], bins = filter_values)
    counts = pd.value_counts(out)
    
    #Counts
    tmp_counts = []
    for i in range(0, len(counts)):
        index = counts.index[i]
        tmp_counts.append((int(index[1:index.find(',')]), counts[i]))
        
    
    tmp_counts = sorted(tmp_counts, key=itemgetter(0))
    counts_index = list(x[0] for x in tmp_counts)
    counts_value = list(x[1] for x in tmp_counts)
    counts = Series(counts_value, index=counts_index)
    COUNTS_VALUE.append(counts)
```
