# Load JSON Data and Convert Them into DataFrame

```python
import pandas as pd
from pandas import Series, DataFrame
import json
```

```python
TEDX_JSON_FILE = "final_tedx.json"
tedx_video_list = []
with open(TEDX_JSON_FILE, "r") as tedx_json_file:
    for line in tedx_json_file:
        tedx_video_list.append(json.loads(line))
    tedx_df = DataFrame(tedx_video_list)
tedx_df.set_index('id', inplace=True, drop=True)
tedx_df
```

# Statistics by Language

## Number of Videos by Language

```python
tedx_df['lang'].value_counts()
```

## Number of View Counts by Language

```python
tedx_df['viewCount'] = tedx_df['viewCount'].fillna(0)
tedx_df[tedx_df.viewCount!='-'][['viewCount', 'lang']].groupby('lang').sum().sort('viewCount', ascending=0)
```

## Number of Events by Language

```python
tedx_df.groupby('lang').event.nunique().order(ascending=False)
```

# Statistics by Country

## Number of Videos by Country

```python
tedx_df['country'].value_counts()
```

## Number of View Counts by Country

```python
tedx_df[tedx_df.viewCount!='-'][['viewCount', 'country']].groupby('country').sum().sort('viewCount', ascending=0)
```

## Number of Events by Country

```python
tedx_df.groupby('country').event.nunique().order(ascending=False)
```

# Statistics by Event

## Number of Videos by Event

```python
tedx_df['event'].value_counts()
```

## Number of View Counts by Event

```python
tedx_df[tedx_df.viewCount!='-'][['viewCount', 'event']].groupby('event').sum().sort('viewCount', ascending=0)[:20]
```
